```python
#load libraries
import numpy as np
from matplotlib import pyplot as plt 
import pandas as pd
from matplotlib.pyplot import figure
import re
```

The following functions define and execute the discrete map
\begin{equation}
\theta_{n+1/2}=\theta_n+\frac{1}{2}K p_n
\end{equation}

\begin{equation}
p_{n+1} = p_n - K \sin(\theta_{n+1/2})
\end{equation}

\begin{equation}
\theta_{n+1}= \theta_{n+1/2}+\frac{1}{2}Kp_{n+1}
\end{equation}

```python
def kick_rotate(p_n=float,theta_n=float,K=float):
    theta_nphalf = theta_n + 1/2*K*p_n
    p_np1 = p_n - K*np.sin(theta_nphalf)
    theta_np1 = theta_nphalf + 1/2*K*p_np1
    return [p_np1,theta_np1]

def propagate(p0=float,theta_0=float,K=float,N=int):
    coords=[[p0,theta_0]]
    for i in range(N):
        new_coords = kick_rotate(p_n=coords[-1][0],theta_n=coords[-1][1],K=K)
        coords.append(new_coords)
    coords=np.array(coords)
    return coords
```

for $N=1000$ being the number of steps, the following plot shows the iterated map of the phase-space development for a set of initial coordinates $\theta_0 \in (-\pi, \pi)$, $p_0 \in (-\pi, \pi)$  
# K=0.005

```python
K=0.005

N=1000
p0 = np.pi/4
theta_0 = np.pi/4

thetas = np.linspace(-np.pi,np.pi,10)
p0s = np.linspace(-np.pi,np.pi,10)

figure(figsize=(12, 4), dpi=100)
for theta_0 in thetas:
    for p0 in p0s:
        coords = propagate(p0=p0,theta_0=theta_0,K=K,N=N)
        plt.plot(coords[:,1]/(np.pi),coords[:,0],linewidth=1)
for x in np.linspace(-3,3,4):
    plt.axvline(x=x,linestyle='--',linewidth=1,color='black')
plt.xlim(-4,4)
plt.xlabel(r'$\theta/\pi$')
plt.ylabel('p')
```
